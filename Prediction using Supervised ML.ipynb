**Predicting the percentage of a students based on no. of study hours.**

**This is a simple linear regression as it involves just two variables.**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

***Reading data from remote link**

```python
data = pd.read_csv("http://bit.ly/w-data")
```

```python
data
```

```python
data.head(5)
```

***Plotting the distridution of scores***

```python
data.plot(subplots = True)
```

***We can see that the above graph is correlated***

```python
data.plot(x = 'Hours', y = 'Scores', style = '*')

plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage scores')

plt.show()
```

***Preparing the dataset*** 

```python
#Splitting depentdant and independant variable

x_train = data.iloc[ : , :-1].values
y_train= data.iloc[ : ,1 ].values

from sklearn.model_selection import train_test_split

#Using train_test_split to split x & y into training the test set

x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size = 0.2, random_state = 0)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

***Training the ML model***

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()

regressor.fit(x_train, y_train)
```

```python
print("Training Completed")
```

***Training the Linear Regresion line**

```python
line = regressor.coef_*x_train + regressor.intercept_

#Plotting the next data

plt.scatter(x_train, y_train)
plt.plot(x_train, line)
plt.show()
```

***Now we can make predication***

```python

print(x_test)

y_pred = regressor.predict(x_test)
```

```python
#Camparing the actual model and predicted model

df = pd.DataFrame({"Actual":y_test, "Predicated":y_pred})

```

```python
df
```

***We can also test by our own data***

```python
hours = [[9.25]]

own_pred = regressor.predict(hours)
```

```python
print("No. of Houres = ", format(hours))

print("Predicated scores = ", format(own_pred[0]))
```

***Evalution of model***

```python

from sklearn import metrics

print("Mean Absolute Error: ",
     metrics.mean_absolute_error(y_test, y_pred))
```
