```python
# Importing basic libraries.

from sklearn import datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# From the given 'iris' dataset, predict the optimum numbers of clusters and represent it visually.

# load the dataset of iris

iris = datasets.load_iris()

# for checking the features of iris datset
iris.feature_names
```

```python
iris
```

```python
# Changes iris dataset into frame using pandas

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
# See the first 5 rows

iris_df.head(5)
```

```python
# Define your target and predictors.

# Take only two attributes...for experimenting with K-Means
X = iris.data[:, :2]
y = iris.target

# check the shape of dataset
X.shape, y.shape 
```

```python
# Let’s have a look at our data through a scatter plot.

plt.scatter(X[:,0], X[:,1], c = y, cmap = 'gist_rainbow')
plt.xlabel('Sepa1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
```

```python
# Finding the optimum number of clusters for k-means classification

x = iris_df.iloc[:, [0, 1, 2, 3]].values
```

```python
x
```

```python
# Importing unsupervised ML model

from sklearn.cluster import KMeans
```

```python
wcss = []


for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting the results onto a line graph, allowing us to observe 'The elbow'.

plt.plot(range(1, 11), wcss)
plt.title('The elbow method')
plt.xlabel('Number of clusters')

# Within cluster sum of squares
plt.ylabel('WCSS') 
```

***We can clearly see why it is called 'The elbow method' from the above graph, the optimum clusters is where the elbow occurs. This is when sum of squares doesn't decrease significantly with every iteration, from this we can choose the number of clusters as 3.***

```python
KMeans
```

```python
# Applying K-means to the dataset / Creating the K-means classifier.
# We are going to use three clusters and a random state of 21.

km = KMeans(n_clusters = 3, random_state=21)
km.fit(X)
```

```python
# Identify the center points of the data.
centers = km.cluster_centers_
print(centers)
```

**Evaluate the results by comparing the original labels/targets**

```python
#This will tell us to which cluster does the data observations belong.

new_labels = km.labels_
```

```python
# Plotted the identified clusters and compare with the answer.

fig, axes = plt.subplots(1, 2, figsize = (16,8))
axes[0].scatter(X[:, 0], X[:, 1], c = y, cmap = 'gist_rainbow',
edgecolor='k', s = 150)
axes[1].scatter(X[:, 0], X[:, 1], c = new_labels, cmap ='jet',
edgecolor = 'k', s = 150)
axes[0].set_xlabel('Sepal length', fontsize=18)
axes[0].set_ylabel('Sepal width', fontsize=18)
axes[1].set_xlabel('Sepal length', fontsize=18)
axes[1].set_ylabel('Sepal width', fontsize=18)
axes[0].tick_params(direction = 'in', length = 10, width = 5, colors = 'k', labelsize = 20)
axes[1].tick_params(direction = 'in', length = 10, width = 5, colors = 'k', labelsize = 20)
axes[0].set_title('Actual', fontsize = 18)
axes[1].set_title('Predicted', fontsize = 18)
```
